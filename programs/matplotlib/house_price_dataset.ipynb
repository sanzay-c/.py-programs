```python
# Remove rows where the sale price has missing (NaN) values
df = df.dropna(subset=['SalePrice'])
```

```python
# The code identifies all the columns in the DataFrame that contain numbers (either floating-point or integers).
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
# It replaces any missing values (NaN) in those numerical columns with the average value of that column.
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())
```

```python
#---object data type in pandas typically refers to columns that store categorical data, 
#     such as strings (text) or mixed types (like dates). 
#     So, when you use include=['object'], you're asking pandas to select all columns that 
#     contain categorical data (i.e., columns that are not numeric, like strings, dates, or categories).

# Select the categorical columns (those with 'object' type)
categorical_columns = df.select_dtypes(include=['object']).columns

# Replace missing values in the categorical columns with the mode
# The .mode() calculates the mode of a column, which is the most frequent (most common) value in that column.
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
# corelation matrix
# normalize data
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
```

```python
print(corr_matrix)
```

```python
plt.figure(figsize=(10, 6)) # here, figsize=(width, height)
plt.title('Correlation Matrix')
sns.heatmap(data=corr_matrix)
plt.show()
```

```python
# covariance matrix
cov_matrix = df.select_dtypes(include=['float64', 'int64']).cov()
print(cov_matrix)
```

```python
plt.figure(figsize=(10, 6)) # here figsize=(width, height)
plt.title('Covariance Matrix')
sns.heatmap(data=cov_matrix)
plt.show()
```
